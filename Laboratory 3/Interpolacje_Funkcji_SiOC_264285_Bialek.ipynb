#Zadanie 1 - Interpolacja wybranych funkcji: sin(x),sin(x^-1),sgn(sin(8x)) przy pomocy jedno-wymiarowej konwolucji korzystając z różnych jąder konwolucji.

```python
import numpy as np
import matplotlib.pyplot as plt

#interpolacja, czyli ma być więcej punktów wynikowych

# Definiowanie funkcji jądrowej h1(x) - Box kernel
def h1(x):
    return np.where((x >= 0) & (x < 1), 1, 0)

# Definiowanie funkcji jądrowej h2(x) - Box kernel
def h2(x):
    return np.where((-1/2 <= x) & (x < 1/2), 1, 0)
# Definiowanie funkcji jądrowej h3(x) - Triangle kernel
def h3(x):
    return np.where((-1 <= x) & (x <= 1), 1 - np.abs(x), 0)


#Definowanie zakresów wraz z ilościami punktów
x1 = np.linspace(-np.pi,np.pi,100)
x2 = np.linspace(-np.pi,np.pi,200)
x3 = np.linspace(-np.pi,np.pi,400)
x4 = np.linspace(-np.pi,np.pi,1000)
# Generowanie funkcji sinus(x) w zakresie od -pi do pi dla różnych ilości punków
y1_1 = np.sin(x1)
y1_2 = np.sin(x2)
y1_3 = np.sin(x3)
y1_4 = np.sin(x4)
# Generowanie funkcji sinus(x^-1) w zakresie od -pi do pi dla różnych ilości punków
y2_1 = np.sin(x1**(-1))
y2_2 = np.sin(x2**(-1))
y2_3 = np.sin(x3**(-1))
y2_4 = np.sin(x4**(-1))
# Generowanie funkcji sign(sin(8x)) w zakresie od -pi do pi dla różnych ilości punków
y3_1 = np.sign(np.sin(8*x1))
y3_2 = np.sign(np.sin(8*x2))
y3_3 = np.sign(np.sin(8*x3))
y3_4 = np.sign(np.sin(8*x4))


# Definiowanie jądra oraz wykonanie jedno-wymiarowej konwolucji
# np.linspace(0, 1, kernel_size) - powoduje równe rozmieszczenie 10 elementów (czyli wielkości kernela) w zakresie od 0 do 1
kernel_size = 10
kernel1= h1(np.linspace(0, 1, kernel_size))
kernel2= h2(np.linspace(0, 1, kernel_size))
kernel3= h3(np.linspace(0, 1, kernel_size))

```

```python
from sklearn.metrics import mean_squared_error
import numpy as np


def calculate_mse(y_true, y_convolved):
    min_len = min(len(y_true), len(y_convolved))
    y_true_adjusted = y_true[:min_len]
    y_convolved_adjusted = np.resize(y_convolved, min_len)

    mse = mean_squared_error(y_true_adjusted, y_convolved_adjusted)*100
    print(mse)

# def calculate_rmse(y_true, y_convolved):
#     min_len = min(len(y_true), len(y_convolved))
#     y_true_adjusted = y_true[:min_len]
#     y_convolved_adjusted = np.resize(y_convolved, min_len)

#     rmse = np.sqrt(mean_squared_error(y_true_adjusted, y_pred))*100
#     print(rmse)




print("sin(x) fun: ")
calculate_mse(y1_1,y_convolved1_1)
calculate_mse(y1_2,y_convolved1_2)
calculate_mse(y1_3,y_convolved1_3)
print("sin(x^-1) fun: ")
calculate_mse(y2_1,y_convolved2_1)
calculate_mse(y2_2,y_convolved2_2)
calculate_mse(y2_3,y_convolved2_3)
print("sign(sin(8x)) fun: ")
calculate_mse(y3_1,y_convolved3_1)
calculate_mse(y3_2,y_convolved3_2)
calculate_mse(y3_3,y_convolved3_3)

```

Na podstawie poniższej strony znalazłem możliwość wygenreowania losowych punktów z rozkładu normlanego - zdefinowanie randomNumberGenerator(rng)

DOC: https://realpython.com/numpy-random-normal/

```python
# Inicjalizacja generatora liczb losowych
rng = np.random.default_rng()
# N=100
# mean=0;
# std=np.pi / 2

# Zdefiniowanie oryginalnych punktów
N = 100
x_original = np.linspace(-np.pi, np.pi, N)
y_original = sin_function(x_original)

def sin_function(x):
    return np.sin(x)


    # Funkcja do interpolacji za pomocą konwolucji
def interpolate_convolution(x, y, factor):
    kernel_size = int(np.ceil(factor))
    kernel = np.ones(kernel_size) / kernel_size

    x_extended = np.linspace(x[0], x[-1], factor * len(x) - (factor - 1))
    y_extended = np.zeros_like(x_extended)
    y_extended[::factor] = y

    y_interpolated = np.convolve(y_extended, kernel, mode='same')

    return x_extended, y_interpolated


interpolated_uniform = interpolate_convolution(x_original, y_original, 4)
#---------------------------------------------------------------------------------


# Funkcja do generowania punktów z rozkładu normalnego przy użyciu  generatora
def generate_normal_points_new(N, mean, std, a, b):
    points = rng.normal(mean, std, N)
    return np.clip(points, a, b)

# Generowanie punktów z rozkładu normalnego
x_normal_new = generate_normal_points_new(100, 0, np.pi / 2, -np.pi, np.pi)
y_normal_new = sin_function(x_normal_new)

# Sortowanie punktów
sorted_indices_new = np.argsort(x_normal_new)
x_normal_sorted_new = x_normal_new[sorted_indices_new]
y_normal_sorted_new = y_normal_new[sorted_indices_new]

# Interpolacja dla punktów z rozkładu normalnego
interpolated_normal_new = interpolate_convolution(x_normal_sorted_new, y_normal_sorted_new, 4)

# Rysowanie wykresów
plt.figure(figsize=(15, 6))

# Wykres dla równomiernie rozłożonych punktów
plt.subplot(1, 2, 1)
plt.plot(x_original, y_original, 'o', label='Original Points')
plt.plot(interpolated_uniform[0], interpolated_uniform[1], label='Interpolated (Uniform)', linestyle='--')
plt.title("Uniformly Distributed Points")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()

# Wykres dla punktów z rozkładu normalnego
plt.subplot(1, 2, 2)
plt.plot(x_normal_sorted_new, y_normal_sorted_new, 'o', label='Normal Distribution Points')
plt.plot(interpolated_normal_new[0], interpolated_normal_new[1], label='Interpolated (Normal)', linestyle='--')
plt.title("Normally Distributed Points")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()

plt.tight_layout()

#---------------------------------------------------------------------------------


from sklearn.metrics import mean_squared_error
import numpy as np


def calculate_mse(y_UNIFROMLY, y_DISTRIBUTED): # true, convolved
    min_len = min(len(y_UNIFROMLY), len(y_DISTRIBUTED))
    y_UNIFORMLY_adjusted = y_UNIFROMLY[:min_len]
    y_DISTRIBUTED_adjusted = np.resize(y_DISTRIBUTED, min_len)
    mse = mean_squared_error(y_UNIFORMLY_adjusted, y_DISTRIBUTED_adjusted)*100
    print(mse)

calculate_mse(interpolated_uniform[1],interpolated_normal_new[1])


```
