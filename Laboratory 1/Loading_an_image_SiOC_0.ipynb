#Loading multiple files at once using glob library

```python
import glob
import cv2
from google.colab.patches import cv2_imshow


path = "/content/image_processing_source/multiple_files/*"

for file in glob.glob(path):
  print(file)
  #Converting files to  numpy array
  arr = cv2.imread(file)
  #Displaying
  cv2_imshow(cv2.resize(arr, (640,426)))
```

#Pillow basic image processing

#Image RGB  to HSV


```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
from skimage import io
from skimage.color import rgb2hsv
from skimage.color import rgb2gray

#Uploadin an image
rgb_img = io.imread("/content/sample_data/Panda.jpg")
#Converting to hsv (hue, saturation, value)
hsv_img = rgb2hsv(rgb_img)
#Converting to hue
hue_img = hsv_img[:, :, 0]
#Converting to value channel
value_img = hsv_img[:, :, 2]
#Converting to grayscale
img_gray = rgb2gray(rgb_img)

fig, (ax0, ax1, ax2,ax3) = plt.subplots(ncols=4, figsize=(8, 2))


ax0.imshow(rgb_img)
ax0.set_title("RGB")
ax0.axis("off")

ax1.imshow(hsv_img)
ax1.set_title("HSV")
ax1.axis("off")

ax2.imshow(hue_img)
ax2.set_title("HUE")
ax2.axis("off")

ax3.imshow(img_gray)
ax3.set_title("GRAYSCALE")
ax3.axis("off")
```
